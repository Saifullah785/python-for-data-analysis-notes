#**JSON Data**

```python
# Define a multi-line JSON string
obj = """
{"name": "Wes",
 "places_lived": ["United States", "Spain", "Germany"],
 "pet": null,
 "siblings": [{"name": "Scott", "age": 30, "pets": ["Zeus", "Zuko"]},
 {"name": "Katie", "age": 38,
 "pets": ["Sixes", "Stache", "Cisco"]}]
}
"""
```

```python
# Import the json module for working with JSON data
import json
# Parse the JSON string into a Python dictionary
result = json.loads(obj)
# Display the resulting dictionary
result
```

```python
# Convert the Python dictionary back into a JSON string
asjson = json.dumps(result)
# Display the resulting JSON string
asjson
```

```python
# Import the pandas library
import pandas as pd
# Create a DataFrame from the 'siblings' list of dictionaries in the JSON data
# The columns of the DataFrame are specified as "name" and "age"
siblings = pd.DataFrame(result["siblings"], columns=["name", "age"])
# Display the DataFrame
siblings
```

#**XML and HTML: Web Scraping**

```python
# Import the pandas and numpy libraries
import pandas as pd
import numpy as np
```

```python
# Define the URL for an HTML file containing a table of failed banks
url = 'https://raw.githubusercontent.com/wesm/pydata-book/d86d8056fa17b180e2a7dc1d9ae1cfb830967d50/examples/fdic_failed_bank_list.html'
# Read all tables from the HTML file into a list of DataFrames
tables = pd.read_html(url)
# Print the number of tables found
len(tables)
# Select the first table from the list
failures = tables[0]
# Display the first few rows of the DataFrame
failures.head()
```

```python
# Convert the 'Closing Date' column to datetime objects
close_timestamps = pd.to_datetime(failures["Closing Date"])
# Count the number of bank failures per year and display the result
close_timestamps.dt.year.value_counts()
```

```python
# Import the objectify module from the lxml library for parsing XML
from lxml import objectify
# Import the urlopen function for opening URLs
from urllib.request import urlopen
# Define the URL for an XML file containing MTA performance data
url = 'https://raw.githubusercontent.com/wesm/pydata-book/d86d8056fa17b180e2a7dc1d9ae1cfb830967d50/datasets/mta_perf/Performance_MNR.xml'
# Open the URL and parse the XML data
with urlopen(url) as f:
    parsed = objectify.parse(f)
# Get the root element of the parsed XML tree
root = parsed.getroot()
```

```python
# Create an empty list to store the data
data = []

# Define a list of fields to skip during parsing
skip_fields = ["PARENT_SEQ", "INDICATOR_SEQ",
               "DESIRED_CHANGE", "DECIMAL_PLACES"]

# Iterate over each 'INDICATOR' element in the XML root
for elt in root.INDICATOR:
    # Create an empty dictionary to store the data for each element
    el_data = {}
    # Iterate over the children of the 'INDICATOR' element
    for child in elt.getchildren():
        # If the child's tag is in the skip_fields list, skip it
        if child.tag in skip_fields:
            continue
        # Otherwise, add the child's tag and its Python value to the dictionary
        el_data[child.tag] = child.pyval
    # Append the dictionary to the data list
    data.append(el_data)
```

```python
# Create a pandas DataFrame from the list of dictionaries
perf = pd.DataFrame(data)
# Display the first few rows of the DataFrame
perf.head()
#
```
