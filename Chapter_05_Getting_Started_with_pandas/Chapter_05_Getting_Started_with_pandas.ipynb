```python
# Create a pandas Series with default index
obj  = pd.Series([4,7,-5,3])
obj
```

```python
# Access the underlying NumPy array of the Series
obj.array
```

```python
# Access the index of the Series
obj.index
```

```python
# Create a pandas Series with a custom index
obj2 = pd.Series([4,7,-5,3], index=['d','b','a','c'])
obj2
```

```python
# Access the custom index of the Series
obj2.index
```

```python
# Access an element of the Series using its custom index
obj2['a']
```

```python
# Modify an element of the Series using its custom index
obj2['d'] = 6
obj2
```

```python
# Access multiple elements of the Series using a list of custom indices
obj2[['c','a','d']]
```

```python
# Filter Series elements based on a boolean condition
obj2[obj2 > 0]
```

```python
# Perform scalar multiplication on the Series
obj2 * 2
```

```python
# Apply a NumPy function to the Series (exponential function)
import numpy as np
np.exp(obj2)
```

```python
# Check for index membership in the Series
'b' in obj2
'e' in obj2
```

```python
# Create a DataFrame from a dictionary of lists
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
        'year': [2000, 2001, 2002, 2001, 2002, 2003],
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
frame = pd.DataFrame(data)
```

```python
# Display the entire DataFrame
frame
```

```python
# Display the first 5 rows of the DataFrame
frame.head()
```

```python
# Display the last 5 rows of the DataFrame
frame.tail()
```

```python
# Create a DataFrame with specified column order
pd.DataFrame(data, columns=['year', 'state', 'pop'])
```

```python
# Create a DataFrame with an additional column ('debt') that has missing values
frame2 = pd.DataFrame(data, columns=['year', 'state', 'pop', 'debt'])
frame2
```

```python
# Access the columns of the DataFrame
frame2.columns
```

```python
# Access a row of the DataFrame by its integer index (label-based)
frame2.loc[1]
```

```python
# Access another row of the DataFrame by its integer index (label-based)
frame2.loc[2]
```

```python
# Assign a scalar value to an entire column in the DataFrame
frame2['debt'] = 16.5
frame2
```

```python
# Assign a NumPy array to a column in the DataFrame
frame2['debt'] = np.arange(6.)
frame2
```

```python
# Assign a pandas Series to a column in the DataFrame (aligns based on index)
val = pd.Series([-1.2, -1.5, -1.7], index=['two', 'four', 'five'])
frame2['debt'] = val
frame2
```

```python
# Create a Series with a custom index and floating-point values
obj =pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
obj
```

```python
# Reindex the Series with a new index, introducing NaN for missing values
obj2 = obj.reindex(['a','b','c','d','e'])
obj2
```

```python
# Create a Series with a sparse index
obj3  = pd.Series(['blue', 'purple', 'yellow'], index=[0, 2, 4])
obj3
```

```python
# Reindex the Series with a full range and forward fill missing values
obj3.reindex(range(6), method='ffill')
```

```python
# Create a DataFrame from a NumPy array with custom index and columns
frame = pd.DataFrame(np.arange(9).reshape((3, 3)),
                     index=['a', 'c', 'd'],
                     columns=['Ohio', 'Texas', 'California'])
frame
```

```python
# Reindex the DataFrame with a new index, introducing NaN for missing rows
frame2 = frame.reindex(index=['a', 'b', 'c', 'd'])
frame2
```

```python
# Reindex the DataFrame with a new set of columns, introducing NaN for missing columns
states = ['Texas', 'Utah', 'California']
frame.reindex(columns=states)
```

```python
# Reindex the DataFrame specifying the axis for reindexing (columns)
frame.reindex(states, axis ='columns')
```

```python
# Select rows and columns from the DataFrame using lists of labels
frame.loc[['a','d','c'],['California','Texas']]
```

```python
# Create a Series of floating-point numbers with a custom index
obj = pd.Series(np.arange(5.), index=['a', 'b', 'c', 'd', 'e'])
obj
```

```python
# Drop a single element from the Series by its index
new_obj = obj.drop('c')
new_obj
```

```python
# Drop multiple elements from the Series by their indices
obj.drop(['d','c'])
```
