## **Arithmetic methods with fill values**

```python
# Create two DataFrames with overlapping but not identical indices and columns
df1 = pd.DataFrame(np.arange(12.).reshape((3, 4)),
                   columns=list('abcd'))
df2 = pd.DataFrame(np.arange(20.).reshape((4, 5)),
                   columns=list('abcde'))
# Introduce a missing value in df2
df2.loc[1, 'b'] = np.nan
df1
df2
```

```python
# Add the two DataFrames (aligns based on both index and column labels, resulting in NaN for non-matching labels and existing NaNs)
df1 + df2
```

```python
# Add the two DataFrames using the add method and fill missing values with 0
df1.add(df2, fill_value=0)
```

```python
# Perform element-wise division (reciprocal) and then demonstrate reverse division using rdiv
1 / df1
df1.rdiv(1)
```

```python
# Reindex df1 to match the columns of df2, filling new columns with 0
df1.reindex(columns=df2.columns, fill_value=0)
```

## **Operations between DataFrame and Series**


```python
# Create a NumPy array
arr = np.arange(12.).reshape((3, 4))
arr
```

```python
# Access the first row of the NumPy array
arr[0]
```

```python
# Subtract the first row from each row of the NumPy array (broadcasting)
arr - arr[0]
```

```python
# Create a DataFrame and extract the first row as a Series
frame = pd.DataFrame(np.arange(12.).reshape((4, 3)),
                     columns=list('bde'),
                     index=['Utah', 'Ohio', 'Texas', 'Oregon'])
series = frame.iloc[0]
frame
series
```

```python
# Subtract the Series (first row) from the DataFrame (aligns by column labels and broadcasts over rows)
frame - series
```

```python
# Create another Series with partially overlapping index with the DataFrame
series2 = pd.Series(range(3), index=['b', 'e', 'f'])
series2
```

```python
# Add the Series to the DataFrame (aligns by column labels, broadcasting over rows, resulting in NaN for non-matching columns)
frame + series2
```

```python
# Extract a column from the DataFrame as a Series
series3 = frame['d']
frame
series3
```

```python
# Subtract the Series (a column) from the DataFrame along the index axis (aligns by index labels and broadcasts over columns)
frame.sub(series3, axis='index')
```

## **Function Application and Mapping**


```python
# Create a DataFrame with random standard normal values
frame = pd.DataFrame(np.random.standard_normal((4, 3)),
                     columns=list("bde"),
                     index=["Utah", "Ohio", "Texas", "Oregon"])
frame
```

```python
# Create a DataFrame with random normal values (redundant code)
frame = pd.DataFrame(np.random.randn(4, 3), columns=list('bde'),
                     index=['Utah', 'Ohio', 'Texas', 'Oregon'])
frame
```

```python
# Calculate the absolute value of each element in the DataFrame using NumPy
np.abs(frame)
```

```python
# Define a function to calculate the range (max - min)
def f1(x):
  return x.max() - x.min()
# Apply the function to each column of the DataFrame
frame.apply(f1)
```

```python
# Apply the function to each row of the DataFrame
frame.apply(f1, axis='columns')
```

```python
# Define a function to return a Series with min and max values
def f2(x):
  return pd.Series([x.min(), x.max()], index=['min', 'max'])
# Apply the function to each column of the DataFrame
frame.apply(f2)
```

```python
# Define a formatting function
def my_format(x):
  return f'{x: .2f}'

# Apply the formatting function to each element of the DataFrame using applymap (deprecated)
frame.applymap(my_format)
```

```python
# Apply the formatting function to each element of a specific Series (column of the DataFrame) using map
frame['e'].map(my_format)
```
