```python
# Create a pandas Series with duplicate values
obj = pd.Series(['c', 'a', 'd', 'a', 'a', 'b', 'b', 'c', 'c'])
```

```python
# Get the unique values from the Series
uniques = obj.unique()
uniques
```

```python
# Count the occurrences of each unique value in the Series
obj.value_counts()
```

```python
# Count the occurrences of each unique value in the underlying NumPy array, without sorting
pd.value_counts(obj.to_numpy(), sort=False)
```

```python
# Display the Series
obj
# Create a boolean mask indicating which elements are in the list ['b', 'c']
mask = obj.isin(['b', 'c'])
# Display the boolean mask
mask
```

```python
# Filter the Series using the boolean mask to show only elements that are 'b' or 'c'
obj[mask]
```

```python
# Create a Series of values to match
to_match = pd.Series(['c', 'a', 'b', 'b', 'c', 'a'])
# Create a Series of unique values
unique_vals = pd.Series(['c', 'b', 'a'])
# Get the integer indices of the unique values that correspond to the elements in to_match
indices = pd.Index(unique_vals).get_indexer(to_match)
# Display the resulting indices
indices
```

```python
# Create a DataFrame with sample data
data = pd.DataFrame({'Qu1': [1, 3, 4, 3, 4],
                     'Qu2': [2, 3, 1, 2, 3],
                     'Qu3': [1, 5, 2, 4, 4]})
# Display the DataFrame
data
```

```python
# Count the occurrences of each unique value in the 'Qu1' column and sort the results by index
data['Qu1'].value_counts().sort_index()
```

```python
# Apply the value_counts function to each column of the DataFrame,
# then fill any missing values (due to a value not appearing in a column) with 0
result = data.apply(pd.value_counts).fillna(0)
# Display the resulting DataFrame
result
```
